```python
from PIL import Image
from diffusers import StableDiffusionPipeline
import torch

# Loading th model
original_image = Image.open("/content/assignment.jpg")

# new size (1280x1280) by adding 128 pixels on each side
new_size = (original_image.width + 256, original_image.height + 256)

# Create a new blank canvas
new_canvas = Image.new("RGB", new_size, (255, 255, 255))

# Pasting the original image
new_canvas.paste(original_image, (128, 128))

# Loading the Stable Diffusion pipeline
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16).to("cuda")

# Prompt giving
prompt = "A young boy with brown hair and a jacket is at the helm of a ship, pointing towards a whimsical island that floats in the sky. The island is teeming with various animals and fantastical creatures. A castle sits atop the island. The boy's eyes are wide with wonder as he gazes up at the magical sight before him."

# Generating the outpainted image
outpainted_image = pipe(prompt=prompt, image=new_canvas, num_inference_steps=100,guidance_scale=13).images[0]

# Saving the outpainted image
outpainted_image.save("outpainted_image.png")
```
